## 融合检索：结合向量和关键词搜索

在本笔记本中，我实现了一个融合检索系统，该系统将语义向量搜索的优势与基于关键字的 BM25 检索相结合。此方法通过捕获概念相似性和确切的关键字匹配来提高检索质量。

## 为什么 Fusion Retrieval 很重要

传统的 RAG 系统通常仅依赖于向量搜索，但这有局限性：

- 向量搜索擅长语义相似性，但可能会错过完全匹配的关键字
- 关键字搜索非常适合特定术语，但缺乏语义理解
- 不同的查询使用不同的检索方法表现得更好

Fusion 检索通过以下方式为我们提供了两全其美的效果：

- 执行基于向量和基于关键字的检索
- 对每种方法的分数进行标准化
- 将它们与加权公式组合
- 根据综合分数对文档进行排名
